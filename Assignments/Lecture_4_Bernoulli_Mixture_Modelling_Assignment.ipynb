### $\textbf{Stochastic Bernoulli Mixture}$

```python
class StochasticBernoulliMixture:
    def __init__(self, n_components, max_iter, tol=1e-3, step_size=1):
        self.n_components = n_components
        self.max_iter = max_iter
        self.tol = tol
        self.step_size = step_size

    def fit(self, x):
        self.x = x
        self.init_params()
        log_bernoullis = self.get_log_bernoullis(self.x)
        self.old_logL = self.get_log_likelihood(log_bernoullis)
        self.Q = 0

        self.approximated_distributions = []
        for _ in range(self.max_iter):
            approx_dists = []
            for _ in range(self.n_components):
                samples = np.random.uniform(low=0, high=1, size=self.x.shape[0])
                approx_dists.append(samples)
            self.approximated_distributions.append(approx_dists)

        for step in range(self.max_iter):
            if step > 0:
                self.old_logL = self.logL

            # S-step
            z_tilde_k = self.sample_latent_variable(step)

            # SA-step
            gamma_k = self.step_size
            new_Q = self.Q + gamma_k * (np.mean(self.get_log_joint_distribution(log_bernoullis, z_tilde_k, self.mu)) - self.Q)
            self.gamma = self.get_responsibilities(log_bernoullis)
            self.remember_params()

            # M-Step
            self.get_Neff()
            self.get_mu()
            self.get_pi()

            # Compute new log_likelihood:
            log_bernoullis = self.get_log_bernoullis(self.x)
            self.logL = self.get_log_likelihood(log_bernoullis)
            if np.isnan(self.logL):
                self.reset_params()
                print(self.logL)
                break

            self.Q = new_Q

    def reset_params(self):
        self.mu = self.old_mu.copy()
        self.pi = self.old_pi.copy()
        self.gamma = self.old_gamma.copy()
        self.get_Neff()
        log_bernoullis = self.get_log_bernoullis(self.x)
        self.logL = self.get_log_likelihood(log_bernoullis)

    def remember_params(self):
        self.old_mu = self.mu.copy()
        self.old_pi = self.pi.copy()
        self.old_gamma = self.gamma.copy()

    def init_params(self):
        self.n_samples = self.x.shape[0]
        self.n_features = self.x.shape[1]
        self.pi = 1/self.n_components * np.ones(self.n_components)
        self.mu = np.random.RandomState(seed=0).uniform(low=0.25, high=0.75, size=(self.n_components, self.n_features))
        self.normalize_mu()

    def normalize_mu(self):
        sum_over_features = np.sum(self.mu, axis=1)
        for k in range(self.n_components):
            self.mu[k,:] /= sum_over_features[k]

    def sample_latent_variable(self, step):
        z_tilde_k = np.zeros(shape=(self.x.shape[0], self.n_components))
        for k in range(self.n_components):
            z_tilde_k[:, k] = np.random.choice(self.approximated_distributions[step][k], size=self.x.shape[0])
        return z_tilde_k

    def get_responsibilities(self, log_bernoullis):
        gamma = np.zeros(shape=(log_bernoullis.shape[0], self.n_components))
        Z =  logsumexp(np.log(self.pi[None,:]) + log_bernoullis, axis=1)
        for k in range(self.n_components):
            gamma[:, k] = np.exp(np.log(self.pi[k]) + log_bernoullis[:,k] - Z)
        return gamma

    def get_log_bernoullis(self, x):
        log_bernoullis = self.get_save_single(x, self.mu)
        log_bernoullis += self.get_save_single(1-x, 1-self.mu)
        return log_bernoullis

    def get_save_single(self, x, mu):
        epsilon = 1e-15
        mu_place = np.maximum(np.where(np.max(mu, axis=0) <= epsilon, epsilon, mu), epsilon)
        return np.tensordot(x, np.log(mu_place), (1,1))

    def get_Neff(self):
        self.Neff = np.sum(self.gamma, axis=0)

    def get_mu(self):
        self.mu = np.einsum('ik,id -> kd', self.gamma, self.x) / self.Neff[:,None]

    def get_pi(self):
        self.pi = self.Neff / self.n_samples

    def get_log_joint_distribution(self, log_bernoullis, z_tilde_k, mu):
        log_joint_dist = np.zeros(self.x.shape[0])
        for k in range(self.n_components):
            log_joint_dist += z_tilde_k[:, k] * (np.log(self.pi[k]) + log_bernoullis[:, k])
        return log_joint_dist

    def predict(self, x):
        log_bernoullis = self.get_log_bernoullis(x)
        gamma = self.get_responsibilities(log_bernoullis)
        return np.argmax(gamma, axis=1)

    def get_sample_log_likelihood(self, log_bernoullis):
        return logsumexp(np.log(self.pi[None,:]) + log_bernoullis, axis=1)

    def get_log_likelihood(self, log_bernoullis):
        return np.mean(self.get_sample_log_likelihood(log_bernoullis))

    def score(self, x):
        log_bernoullis = self.get_log_bernoullis(x)
        return self.get_log_likelihood(log_bernoullis)

    def score_samples(self, x):
        log_bernoullis = self.get_log_bernoullis(x)
        return self.get_sample_log_likelihood(log_bernoullis)
```

### $\textbf{KModes vs Stochastic Bernoulli Mixture}$

```python
X = df.drop(columns=['class'])
y = df['class']

encoder = OneHotEncoder()
encoded_data = encoder.fit_transform(X)
encoded_df = pd.DataFrame(encoded_data.toarray(), columns=encoder.get_feature_names_out(X.columns))

encoded_df.head()
```

```python
km = KModes(n_clusters=5, init='Huang', n_init=5)
df["km_clusters"] = km.fit_predict(encoded_df)
df["km_clusters"]
```

```python
stochastic_bernoulli = StochasticBernoulliMixture(n_components=5, max_iter=500)

stochastic_bernoulli.fit(encoded_df)
df["sbm_clusters"] = stochastic_bernoulli.predict(encoded_df)
df["sbm_clusters"]
```

```python
pca = PCA(n_components=2)
pca_result = pca.fit_transform(X)

pca_df = pd.DataFrame(data=pca_result, columns=['PC1', 'PC2'])

pca_df['km_cluster'] = df['km_clusters'].astype('category')
pca_df['sbm_cluster'] = df['sbm_clusters'].astype('category')

plt.figure(figsize=(15, 6))

plt.subplot(1, 2, 1)
sns.scatterplot(data=pca_df, x='PC1', y='PC2', hue='km_cluster', s=100, palette='RdBu')
plt.title('KModes Clustering (PCA)')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend(title='Cluster', bbox_to_anchor=(1.05, 1), loc='upper left')

plt.subplot(1, 2, 2)
sns.scatterplot(data=pca_df, x='PC1', y='PC2', hue='sbm_cluster', s=100, palette='RdBu')
plt.title('Stochastic Bernoulli Mixture Clustering (PCA)')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend(title='Cluster', bbox_to_anchor=(1.05, 1), loc='upper left')

plt.tight_layout()
plt.show()
```

```python
fmi_km = fowlkes_mallows_score(y, df['km_clusters'])
ari_km = adjusted_rand_score(y, df['km_clusters'])
nmi_km = normalized_mutual_info_score(y, df['km_clusters'])

fmi_sbm = fowlkes_mallows_score(y, df['sbm_clusters'])
ari_sbm = adjusted_rand_score(y, df['sbm_clusters'])
nmi_sbm = normalized_mutual_info_score(y, df['sbm_clusters'])

results = {
    'Method': ['Kmodes', 'Stochastic Bernoulli Mixture'],
    'FMI': [fmi_km, fmi_sbm],
    'ARI': [ari_km, ari_sbm],
    'NMI': [nmi_km, nmi_sbm]
}

df_results = pd.DataFrame(results)
df_results
```

### $\textbf{Report}$

In this report, we analyze the performance of two clustering algorithms, KModes and Stochastic Bernoulli Mixture, on a dataset containing categorical data. We focus on three performance metrics: Fowlkes-Mallows Index (FMI), Adjusted Rand Index (ARI), and Normalized Mutual Information (NMI). 

### Performance Metrics:
- **Fowlkes-Mallows Index (FMI)**: Measures the similarity between two clusterings by calculating the geometric mean of the pairwise precision and recall.
- **Adjusted Rand Index (ARI)**: Measures the similarity between two clusterings by considering all pairs of samples and counting pairs that are assigned to the same or different clusters in the predicted and true clusterings, adjusted for chance agreement.
- **Normalized Mutual Information (NMI)**: Measures the amount of information shared between the clustering result and the ground truth labels, normalized by the entropy of the clusterings.

#### Fowlkes-Mallows Index (FMI)

We observe that SBM outperforms Kmodes in terms of FMI, with SBM achieving a higher value (0.445841) compared to Kmodes (0.329336). FMI measures the similarity between the true pairwise associations and the pairwise associations in the predicted clusters. The higher FMI score of SBM suggests that SBM captures the true associations more accurately than Kmodes.

#### Adjusted Rand Index (ARI)

Similarly, SBM demonstrates superior performance over Kmodes in terms of ARI. SBM achieves a higher ARI score (0.305205) compared to Kmodes (0.196024). ARI measures the similarity between the true clustering and the clustering results. The higher ARI score of SBM indicates that SBM produces clusterings that are more aligned with the ground truth labels.

#### Normalized Mutual Information (NMI)

SBM also outperforms Kmodes in terms of NMI, with SBM achieving a higher NMI score (0.561921) compared to Kmodes (0.424560). NMI measures the mutual information between the true labels and the predicted clusters, normalized by the entropy of the true labels and the predicted clusters. The higher NMI score of SBM indicates that SBM captures the underlying structure of the data more effectively than Kmodes.

### Conclusion

In conclusion, our comparison reveals that Stochastic Bernoulli Mixture clustering algorithm consistently outperforms Kmodes in terms of FMI, ARI, and NMI metrics. SBM demonstrates superior clustering quality and better alignment with the ground truth labels compared to Kmodes. These findings suggest that SBM is more suitable for clustering binary data and offers better performance in capturing the underlying structure of the dataset.
